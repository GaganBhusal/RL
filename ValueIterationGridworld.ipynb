```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(data):
    fig, ax = plt.subplots()
    ax.axis('off')
    table = ax.table(cellText = data, loc = 'center')
    table.auto_set_font_size(True)
    # table.set_fontsize(12)
    table.scale(1, 2) 
```

```python
reward = -1
action_names = {
    "UP" : (-1, 0),
    "DOWN" :(1, 0),
    "RIGHT" : (0, 1),
    "LEFT" : (0, -1),
}
```

```python
def summ(row, col, data, actions, reward):
    
    states = []
    for a in actions:
        st = [row, col] + a
        a, b = st
        if 0<=a<5 and 0<=b<5:
            st = (a, b)
        else:
            st = (row, col)
        states.append(st)
    return max(reward + data[s] for s in states)
    
```

```python
def evaluate_value(data,reward, action_names):
    data1 = np.zeros((5, 5))
    actions = np.array(list(action_names.values()))
    for row in range(5):
        for col in range(5):
            if (row == 0 and col ==0) or (row == 4 and col == 4):
                continue
            data1[row, col] =  round((summ(row, col, data, actions, reward)), 1)
            
    plot(data1)
    return data1
```

```python
def evaluate_policy(data, action_names):
    data2 = np.zeros((5, 5)).tolist()
    actions = np.array(list(action_names.values()))
    for row in range(5):
        for col in range(5):
            diff = []
            key = ""
            if (row == 0 and col ==0) or (row == 4 and col == 4):
                    continue
            states = []
            for a in actions:
                st = [row, col] + a
                a, b = st
                if 0<=a<5 and 0<=b<5:
                    states.append((a,b))
                else:
                    states.append((row, col))
            for i, s in enumerate(states):
                diff.append(data[row, col] - data[s])
            m = np.min(diff)
            a = 0
            for i, d in enumerate(diff):
                if d == m:
                    a += 1
                    if a==len(states):
                        key = "ALL"
                        continue
                    min_state = tuple(np.array(states[i]) - (row, col))
                    key += "".join(list((filter(lambda key: action_names[key] == min_state, action_names)))) + " "
            data2[row][col] = key
    plot(data2)
    return data2


                    
            
```

```python
def value_iteration(data, reward, action_names):
    data = np.zeros((5, 5))
    delta = 100
    while delta > 0.3:
        data1 = evaluate_value(data, reward, action_names)
        delta = max(data[row][col] - data1[row][col] for col in range(5) for row in range(5))
        data = data1
    evaluate_policy(data, action_names)
```

```python
value_iteration(data, reward, action_names)
```
